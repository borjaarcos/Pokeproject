```python
#Cargo las librerias para poder utilizarlas en los demás ejercicios
import math as m
import numpy as np
import pandas as pd

dat_base = pd.read_csv("pokemon_data.csv")

dat_base.info()

#Miramos a ver si hay valores nulos 
print("Cantidad de nulos por columna:")

# Muestra un booleano para cada fila, True si 'moves' es null
print(dat_base['moves'].isnull())

# Listar todos los registros (filas) donde 'moves' es null
null_moves = dat_base[dat_base['moves'].isnull()]
print(dat_base['stats'])
```

### ¿Hay valores nulos o faltantes?

Es una celda vacía, sin información (NaN en Python).

Puede representar errores de carga, datos no recolectados, o no aplicables.


```python
print("Cantidad de nulos por columna:")
print(dat_base.isnull().sum())
```

```python
# Saber cuántas filas duplicadas hay
duplicados = dat_base.duplicated().sum()
print(f"Número de filas duplicadas: {duplicados}")
```

```python
print("Número de valores únicos por columna:")
print(dat_base.nunique())

print("\nValores únicos Género:")
print(dat_base['types'].unique())

print(dat_base['types'])
```
